```python
import numpy as np

# Simulation Parameters
SIM_TIME = 61  # 60 minutes in steps of 1 minute
ARRIVAL_RATE = 6  # Average workers per minute
ELEVATOR_CAPACITY = 12
FLOORS = {2: 1/3, 3: 1/3, 4: 1/3}
STAIR_PROBABILITY = {2: 0.5, 3: 0.33, 4: 0.1}
ELEVATOR_TIMES = {(0, 2): 1, (0, 3): 1.5, (0, 4): 1.75,
                  (2, 0): 1, (2, 3): 0.5, (2, 4): 0.75,
                  (3, 0): 1.5, (3, 2): 0.5, (3, 4): 0.5,
                  (4, 0): 1.75, (4, 2): 0.75, (4, 3): 0.5}
DOOR_TIME = 0.1

# Data Collection
wait_times = []
stair_usage = {2: 0, 3: 0, 4: 0}
queue_at_830 = None
queue_at_845 = None
queue_at_900 = None
last_boarding_time = 0
elevator_queue = []
total_workers = 0
elevator = []  # Current passengers in the elevator

def simulate():
    global total_workers, last_boarding_time, queue_at_830, queue_at_845, queue_at_900
    for minute in range(SIM_TIME):
        # Worker Arrivals
        arrivals = np.random.poisson(ARRIVAL_RATE)
        for _ in range(arrivals):
            total_workers += 1
            destination = np.random.choice(list(FLOORS.keys()), p=list(FLOORS.values()))
            if np.random.rand() < STAIR_PROBABILITY[destination]:
                stair_usage[destination] += 1
            else:
                elevator_queue.append((minute, destination))
        
        # Print Queue Length Every Minute
        print(f"Time {minute}:00 - Queue Length: {len(elevator_queue)}")
        
        # Queue Tracking (Exact Times)
        if minute == 30:
            queue_at_830 = len(elevator_queue)
        if minute == 45:
            queue_at_845 = len(elevator_queue)
        if minute == 60:
            queue_at_900 = len(elevator_queue)
        
        # Elevator Boarding (Continuous Processing)
        if len(elevator_queue) > 0 and (len(elevator) == 0 or len(elevator) < ELEVATOR_CAPACITY):
            boarding = elevator_queue[:ELEVATOR_CAPACITY - len(elevator)]
            elevator_queue[:] = elevator_queue[len(boarding):]
            wait_times.extend(minute - entry[0] for entry in boarding)
            elevator.extend(boarding)
            last_boarding_time = minute
        
        # Elevator Travel
        if len(elevator) > 0:
            destinations = sorted(set(entry[1] for entry in elevator))
            prev_floor = 0
            travel_time = 0
            for stop in destinations:
                travel_time += ELEVATOR_TIMES.get((prev_floor, stop), 0) + DOOR_TIME
                prev_floor = stop
            elevator.clear()
        

    # Ensure None values default to 0 if no queue was recorded
    queue_at_830 = queue_at_830 if queue_at_830 is not None else 0
    queue_at_845 = queue_at_845 if queue_at_845 is not None else 0
    queue_at_900 = queue_at_900 if queue_at_900 is not None else 0

def visualize_results():
    print("Simulation Results:")
    print(f"Total Workers Evaluated: {total_workers}")
    print(f"Average Wait Time: {np.mean(wait_times) if wait_times else 0:.2f} minutes")
    print(f"Workers Using Stairs: {stair_usage}")
    print(f"Average Number Walking to Floors: 2nd - {stair_usage[2]}, 3rd - {stair_usage[3]}, 4th - {stair_usage[4]}")
    print(f"Time Last Worker Boards: {last_boarding_time} minutes")
    print(f"Queue Length at 8:30 AM: {queue_at_830}")
    print(f"Queue Length at 8:45 AM: {queue_at_845}")
    print(f"Queue Length at 9:00 AM: {queue_at_900}")

if __name__ == "__main__":
    simulate()
    visualize_results()
```

```python
# Updating the simulation code with improvements while adhering to constraints

import numpy as np
import matplotlib.pyplot as plt
import simpy

# Simulation Parameters
SIM_TIME = 60  # 60 minutes (8:00 AM - 9:00 AM)
ARRIVAL_RATE = 6  # Average workers per minute
ELEVATOR_CAPACITY = 12
FLOORS = {2: 1/3, 3: 1/3, 4: 1/3}  # Equal probability of destination floors
STAIR_PROBABILITY = {2: 0.5, 3: 0.33, 4: 0.1}
ELEVATOR_TIMES = {(0, 2): 1.00, (0, 3): 1.5, (0, 4): 1.75,
                  (2, 0): 1.00, (2, 3): 0.50, (2, 4): 0.75,
                  (3, 0): 1.50, (3, 2): 0.50, (3, 4): 0.50,
                  (4, 0): 1.75, (4, 2): 0.75, (4, 3): 0.50}
DOOR_TIME = 0.5

# Data collection
wait_times = []
stair_usage = {2: 0, 3: 0, 4: 0}
queue_sizes = []
idle_time = 0
total_run_time = 0

elevator_queue = []

def worker(env):
    while env.now < SIM_TIME:
        inter_arrival_time = np.random.exponential(1 / ARRIVAL_RATE)
        yield env.timeout(inter_arrival_time)

        destination = np.random.choice(list(FLOORS.keys()), p=list(FLOORS.values()))
        if np.random.rand() < STAIR_PROBABILITY[destination]:
            stair_usage[destination] += 1
            continue

        request_time = env.now
        elevator_queue.append((request_time, destination))

def elevator_process(env):
    global idle_time, total_run_time

    while True:
        if not elevator_queue:
            idle_time += 0.5
            yield env.timeout(0.5)
            continue

        queue_sizes.append(len(elevator_queue))

        yield env.timeout(0.5)  # Small delay to allow queue to build

        # Prioritize longest waiting passengers
        elevator_queue.sort(key=lambda x: x[0])  
        boarding = elevator_queue[:ELEVATOR_CAPACITY]
        elevator_queue[:] = elevator_queue[ELEVATOR_CAPACITY:]

        # Collect wait times
        for entry in boarding:
            wait_times.append(env.now - entry[0])

        destinations = [entry[1] for entry in boarding]
        stops = sorted(set(destinations))
        prev_floor = 0

        for stop in stops:
            if (prev_floor, stop) not in ELEVATOR_TIMES:
                raise ValueError(f"Missing travel time from {prev_floor} to {stop}")

            travel_time = ELEVATOR_TIMES[(prev_floor, stop)]
            yield env.timeout(travel_time + DOOR_TIME)
            total_run_time += travel_time + DOOR_TIME
            prev_floor = stop

# Run simulation
env = simpy.Environment()
env.process(worker(env))
env.process(elevator_process(env))
env.run(until=SIM_TIME)

# Visualization
plt.figure(figsize=(10, 5))
plt.hist(wait_times, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Number of Workers")
plt.title("Distribution of Elevator Wait Times")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(queue_sizes, label="Queue Size Over Time")
plt.xlabel("Time (minutes)")
plt.ylabel("Queue Size")
plt.title("Queue Length Over Time")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(stair_usage.keys(), stair_usage.values(), color='gray', alpha=0.7)
plt.xlabel("Destination Floor")
plt.ylabel("Number of Workers Taking Stairs")
plt.title("Stair Usage per Floor")
plt.show()

# Elevator utilization calculation
utilization = (total_run_time / (SIM_TIME - idle_time)) * 100 if (SIM_TIME - idle_time) > 0 else 0
print(f"Elevator Utilization Rate: {utilization:.2f}%")
print(f"Average Wait Time: {np.mean(wait_times):.2f} minutes")
print(f"Max Queue Length: {max(queue_sizes) if queue_sizes else 0}")
```

```python
import numpy as np
import heapq
from collections import defaultdict
import random

class ElevatorSimulation:
    def __init__(self):
        # Simulation parameters
        self.arrival_rate = 6.0  # workers per minute
        self.arrival_mean = 1 / self.arrival_rate  # mean time between arrivals
        self.elevator_capacity = 12
        self.floors = [0, 2, 3, 4]  # G, 2, 3, 4
        self.floor_probabilities = {2: 1/3, 3: 1/3, 4: 1/3}
        self.walking_probabilities = {2: 0.5, 3: 0.33, 4: 0.1}
        
        # Travel times between floors (in minutes)
        self.travel_times = {
            (0, 2): 1.00, (0, 3): 1.5, (0, 4): 1.75,
            (2, 0): 1.00, (2, 3): 0.50, (2, 4): 0.75,
            (3, 0): 1.50, (3, 2): 0.50, (3, 4): 0.50,
            (4, 0): 1.75, (4, 2): 0.75, (4, 3): 0.50
        }
        self.door_time = 0.5  # time door stays open at each stop
        
        # Simulation state
        self.current_time = 0
        self.elevator_floor = 0
        self.elevator_destination = 0
        self.elevator_available_time = 0
        self.elevator_passengers = []
        self.waiting_queue = []
        self.event_queue = []
        self.walkers = {2: 0, 3: 0, 4: 0}
        self.wait_times = []
        self.queue_sizes = defaultdict(list)
        self.last_boarding_time = 0

    def generate_arrivals(self, start_time=8*60, end_time=9*60):
        """Generate worker arrivals from 8:00 to 9:00."""
        current_time = start_time
        
        while current_time < end_time:
            # Generate next arrival time (exponential distribution)
            interarrival_time = np.random.exponential(self.arrival_mean)
            current_time += interarrival_time
            
            if current_time < end_time:
                # Determine destination floor
                floor = np.random.choice([2, 3, 4], p=[1/3, 1/3, 1/3])
                
                # Add arrival event to queue
                heapq.heappush(self.event_queue, (current_time, 'arrival', {'time': current_time, 'floor': floor}))

    def process_arrival(self, worker):
        """Process a worker arrival."""
        arrival_time = worker['time']
        destination_floor = worker['floor']
        
        # Check if worker decides to walk
        walk_prob = self.walking_probabilities[destination_floor]
        if len(self.waiting_queue) > self.elevator_capacity and random.random() < walk_prob:
            self.walkers[destination_floor] += 1
            return
        
        # Add worker to waiting queue
        self.waiting_queue.append({'arrival_time': arrival_time, 'floor': destination_floor})
        
        # Record queue size for specific times
        self.record_queue_size(arrival_time)
        
        # If elevator is idle at ground floor, dispatch it
        if self.elevator_floor == 0 and self.current_time >= self.elevator_available_time and len(self.waiting_queue) > 0:
            self.dispatch_elevator()

    def dispatch_elevator(self):
        """Load passengers and send elevator to floors."""
        # Load passengers up to capacity
        passengers_to_board = min(len(self.waiting_queue), self.elevator_capacity)
        
        passengers_by_floor = defaultdict(list)
        for _ in range(passengers_to_board):
            passenger = self.waiting_queue.pop(0)
            self.last_boarding_time = self.current_time
            wait_time = self.current_time - passenger['arrival_time']
            self.wait_times.append(wait_time)
            passengers_by_floor[passenger['floor']].append(passenger)
        
        # Determine stops
        stops = sorted(passengers_by_floor.keys())
        
        # Create journey plan
        current_floor = 0
        journey_time = 0
        
        for floor in stops:
            # Add travel time to floor
            journey_time += self.travel_times[(current_floor, floor)]
            # Add door opening time
            journey_time += self.door_time
            current_floor = floor
        
        # Add elevator return journey
        if current_floor != 0:
            journey_time += self.travel_times[(current_floor, 0)]
        
        # Schedule elevator return
        return_time = self.current_time + journey_time
        self.elevator_available_time = return_time
        
        # Schedule next dispatch if there are still people waiting
        if self.waiting_queue:
            heapq.heappush(self.event_queue, (return_time, 'elevator_return', None))

    def record_queue_size(self, time):
        """Record queue size at specific times."""
        # Convert time to hours:minutes for easier comparison
        hours = int(time // 60)
        minutes = int(time % 60)
        
        # Check if we're at one of our target times (8:30, 8:45, 9:00)
        if (hours == 8 and minutes == 30) or (hours == 8 and minutes == 45) or (hours == 9 and minutes == 0):
            target_time = f"{hours}:{minutes:02d}"
            self.queue_sizes[target_time].append(len(self.waiting_queue))

    def run_simulation(self):
        """Run the simulation from 8:00 to 9:00."""
        # Initialize simulation
        self.current_time = 8 * 60  # Start at 8:00 AM
        
        # Generate all arrivals
        self.generate_arrivals()
        
        # Process events
        while self.event_queue:
            time, event_type, data = heapq.heappop(self.event_queue)
            self.current_time = time
            
            if event_type == 'arrival':
                self.process_arrival(data)
            elif event_type == 'elevator_return':
                if self.waiting_queue:
                    self.dispatch_elevator()
        
        # Process any remaining people in queue
        while self.waiting_queue:
            self.dispatch_elevator()
            self.current_time = self.elevator_available_time
            
        # Calculate final statistics
        return self.calculate_statistics()

    def calculate_statistics(self):
        """Calculate and return simulation statistics."""
        stats = {
            'average_wait_time': np.mean(self.wait_times) if self.wait_times else 0,
            'walkers_by_floor': dict(self.walkers),
            'last_boarding_time': self.convert_minutes_to_time(self.last_boarding_time),
            'queue_sizes': dict(self.queue_sizes)
        }
        return stats
    
    def convert_minutes_to_time(self, minutes):
        """Convert minutes since midnight to HH:MM format."""
        hours = int(minutes // 60)
        mins = int(minutes % 60)
        return f"{hours}:{mins:02d}"

# Run multiple simulations and average the results
def run_multiple_simulations(num_simulations=10):
    total_wait_times = 0
    total_walkers = {2: 0, 3: 0, 4: 0}
    last_boarding_times = []
    queue_sizes = {"8:30": [], "8:45": [], "9:00": []}
    
    for _ in range(num_simulations):
        sim = ElevatorSimulation()
        results = sim.run_simulation()
        
        total_wait_times += results['average_wait_time']
        
        for floor, count in results['walkers_by_floor'].items():
            total_walkers[floor] += count
        
        # Convert time back to minutes for averaging
        time_parts = results['last_boarding_time'].split(':')
        minutes = int(time_parts[0]) * 60 + int(time_parts[1])
        last_boarding_times.append(minutes)
        
        # Aggregate queue sizes
        for time, sizes in results['queue_sizes'].items():
            if sizes:  # Only add if there's a value
                queue_sizes[time].extend(sizes)
    
    # Calculate averages
    avg_stats = {
        'average_wait_time': total_wait_times / num_simulations,
        'walkers_by_floor': {floor: count / num_simulations for floor, count in total_walkers.items()},
        'last_boarding_time': sum(last_boarding_times) / num_simulations if last_boarding_times else 0,
        'queue_sizes': {time: sum(sizes) / len(sizes) if sizes else 0 for time, sizes in queue_sizes.items()}
    }
    
    # Convert last boarding time back to HH:MM
    avg_stats['last_boarding_time'] = f"{int(avg_stats['last_boarding_time'] // 60)}:{int(avg_stats['last_boarding_time'] % 60):02d}"
    
    return avg_stats

# Run simulations and print results
if __name__ == "__main__":
    np.random.seed(42)  # For reproducibility
    num_simulations = 30
    print(f"Running {num_simulations} simulations...")
    results = run_multiple_simulations(num_simulations)
    
    print("\nSimulation Results:")
    print(f"Average wait time: {results['average_wait_time']:.2f} minutes")
    print("\nNumber of people who walked to each floor:")
    for floor, count in results['walkers_by_floor'].items():
        print(f"  Floor {floor}: {count:.1f} workers")
    
    print(f"\nTime the last worker boards the elevator: {results['last_boarding_time']}")
    
    print("\nAverage queue size at specific times:")
    for time, size in results['queue_sizes'].items():
        print(f"  {time}: {size:.1f} workers")
```

```python
import numpy as np
import heapq
from collections import defaultdict
import random
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

class ElevatorSimulation:
    def __init__(self):
        # Simulation parameters
        self.arrival_rate = 6.0  # workers per minute
        self.arrival_mean = 1 / self.arrival_rate  # mean time between arrivals
        self.elevator_capacity = 12
        self.floors = [0, 2, 3, 4]  # G, 2, 3, 4
        self.floor_probabilities = {2: 1/3, 3: 1/3, 4: 1/3}
        self.walking_probabilities = {2: 0.5, 3: 0.33, 4: 0.1}
        
        # Travel times between floors (in minutes)
        self.travel_times = {
            (0, 2): 1.00, (0, 3): 1.5, (0, 4): 1.75,
            (2, 0): 1.00, (2, 3): 0.50, (2, 4): 0.75,
            (3, 0): 1.50, (3, 2): 0.50, (3, 4): 0.50,
            (4, 0): 1.75, (4, 2): 0.75, (4, 3): 0.50
        }
        self.door_time = 0.5  # time door stays open at each stop
        
        # Simulation state
        self.current_time = 0
        self.elevator_floor = 0
        self.elevator_destination = 0
        self.elevator_available_time = 0
        self.elevator_passengers = []
        self.waiting_queue = []
        self.event_queue = []
        self.walkers = {2: 0, 3: 0, 4: 0}
        self.wait_times = []
        self.queue_sizes_at_times = {"8:30": [], "8:45": [], "9:00": []}
        self.queue_size_history = []  # Track queue size over time
        self.last_boarding_time = 0
        
        # For visualization
        self.queue_timeline = []
        self.elevator_trips = []
        self.waiting_time_by_arrival = []

    def generate_arrivals(self, start_time=8*60, end_time=9*60+1):
        """Generate worker arrivals from 8:00 to 9:00."""
        current_time = start_time
        
        while current_time < end_time:
            # Generate next arrival time (exponential distribution)
            interarrival_time = np.random.exponential(self.arrival_mean)
            current_time += interarrival_time
            
            if current_time < end_time:
                # Determine destination floor
                floor = np.random.choice([2, 3, 4], p=[1/3, 1/3, 1/3])
                
                # Add arrival event to queue
                heapq.heappush(self.event_queue, (current_time, 'arrival', {'time': current_time, 'floor': floor}))

    def process_arrival(self, worker):
        """Process a worker arrival."""
        arrival_time = worker['time']
        destination_floor = worker['floor']
        
        # Check if worker decides to walk
        walk_prob = self.walking_probabilities[destination_floor]
        if len(self.waiting_queue) > self.elevator_capacity and random.random() < walk_prob:
            self.walkers[destination_floor] += 1
            return
        
        # Add worker to waiting queue
        self.waiting_queue.append({'arrival_time': arrival_time, 'floor': destination_floor})
        
        # Record queue size for timeline
        self.queue_timeline.append((arrival_time, len(self.waiting_queue)))
        
        # Check specific times for queue size monitoring
        self.check_specific_times(arrival_time)
        
        # If elevator is idle at ground floor, dispatch it
        if self.elevator_floor == 0 and self.current_time >= self.elevator_available_time and len(self.waiting_queue) > 0:
            self.dispatch_elevator()

    def check_specific_times(self, current_time):
        """Check if we're at a specific monitoring time."""
        formatted_time = self.convert_minutes_to_time(current_time)
        if formatted_time in self.queue_sizes_at_times and current_time % 15 < 0.5:  # Within 30 seconds of target time
            self.queue_sizes_at_times[formatted_time].append(len(self.waiting_queue))

    def dispatch_elevator(self):
        """Load passengers and send elevator to floors."""
        # Record elevator trip start
        trip_start = self.current_time
        
        # Load passengers up to capacity
        passengers_to_board = min(len(self.waiting_queue), self.elevator_capacity)
        
        passengers_by_floor = defaultdict(list)
        for _ in range(passengers_to_board):
            passenger = self.waiting_queue.pop(0)
            self.last_boarding_time = self.current_time
            wait_time = self.current_time - passenger['arrival_time']
            self.wait_times.append(wait_time)
            self.waiting_time_by_arrival.append((passenger['arrival_time'], wait_time))
            passengers_by_floor[passenger['floor']].append(passenger)
        
        # Record queue size after boarding
        self.queue_timeline.append((self.current_time, len(self.waiting_queue)))
        
        # Determine stops
        stops = sorted(passengers_by_floor.keys())
        
        # Create journey plan
        current_floor = 0
        journey_time = 0
        
        for floor in stops:
            # Add travel time to floor
            journey_time += self.travel_times[(current_floor, floor)]
            # Add door opening time
            journey_time += self.door_time
            current_floor = floor
        
        # Add elevator return journey
        if current_floor != 0:
            journey_time += self.travel_times[(current_floor, 0)]
        
        # Schedule elevator return
        return_time = self.current_time + journey_time
        self.elevator_available_time = return_time
        
        # Record elevator trip
        self.elevator_trips.append((trip_start, return_time, passengers_to_board))
        
        # Schedule next dispatch if there are still people waiting
        if self.waiting_queue:
            heapq.heappush(self.event_queue, (return_time, 'elevator_return', None))

    def run_simulation(self):
        """Run the simulation from 8:00 to 9:00."""
        # Initialize simulation
        self.current_time = 8 * 60  # Start at 8:00 AM
        
        # Generate all arrivals
        self.generate_arrivals()
        
        # Process events
        while self.event_queue:
            time, event_type, data = heapq.heappop(self.event_queue)
            self.current_time = time
            
            if event_type == 'arrival':
                self.process_arrival(data)
            elif event_type == 'elevator_return':
                if self.waiting_queue:
                    self.dispatch_elevator()
        
        # Explicitly check the queue size at exactly 9:00 if not already recorded
        if not self.queue_sizes_at_times["9:00"]:
            self.queue_sizes_at_times["9:00"].append(len(self.waiting_queue))
            self.queue_timeline.append((9*60, len(self.waiting_queue)))
        
        # Process any remaining people in queue
        while self.waiting_queue:
            self.dispatch_elevator()
            self.current_time = self.elevator_available_time
            
        # Calculate final statistics
        return self.calculate_statistics()

    def calculate_statistics(self):
        """Calculate and return simulation statistics."""
        stats = {
            'average_wait_time': np.mean(self.wait_times) if self.wait_times else 0,
            'walkers_by_floor': dict(self.walkers),
            'last_boarding_time': self.convert_minutes_to_time(self.last_boarding_time),
            'queue_sizes_at_times': {time: np.mean(sizes) if sizes else 0 for time, sizes in self.queue_sizes_at_times.items()},
            'queue_timeline': self.queue_timeline,
            'elevator_trips': self.elevator_trips,
            'waiting_time_by_arrival': self.waiting_time_by_arrival
        }
        return stats
    
    def convert_minutes_to_time(self, minutes):
        """Convert minutes since midnight to HH:MM format."""
        hours = int(minutes // 60)
        mins = int(minutes % 60)
        return f"{hours}:{mins:02d}"

# Run multiple simulations and average the results
def run_multiple_simulations(num_simulations=10):
    total_wait_times = 0
    total_walkers = {2: 0, 3: 0, 4: 0}
    last_boarding_times = []
    queue_sizes = {"8:30": [], "8:45": [], "9:00": []}
    
    combined_queue_timeline = []
    combined_elevator_trips = []
    combined_waiting_time = []
    
    for i in range(num_simulations):
        sim = ElevatorSimulation()
        results = sim.run_simulation()
        
        total_wait_times += results['average_wait_time']
        
        for floor, count in results['walkers_by_floor'].items():
            total_walkers[floor] += count
        
        # Convert time back to minutes for averaging
        time_parts = results['last_boarding_time'].split(':')
        minutes = int(time_parts[0]) * 60 + int(time_parts[1])
        last_boarding_times.append(minutes)
        
        # Aggregate queue sizes
        for time, size in results['queue_sizes_at_times'].items():
            queue_sizes[time].append(size)
            
        # Save data for visualization from the first simulation
        if i == 0:
            combined_queue_timeline = results['queue_timeline']
            combined_elevator_trips = results['elevator_trips']
            combined_waiting_time = results['waiting_time_by_arrival']
    
    # Calculate averages
    avg_stats = {
        'average_wait_time': total_wait_times / num_simulations,
        'walkers_by_floor': {floor: count / num_simulations for floor, count in total_walkers.items()},
        'last_boarding_time': sum(last_boarding_times) / num_simulations if last_boarding_times else 0,
        'queue_sizes': {time: sum(sizes) / len(sizes) if sizes else 0 for time, sizes in queue_sizes.items()},
        'queue_timeline': combined_queue_timeline,
        'elevator_trips': combined_elevator_trips,
        'waiting_time_by_arrival': combined_waiting_time
    }
    
    # Convert last boarding time back to HH:MM
    avg_stats['last_boarding_time'] = f"{int(avg_stats['last_boarding_time'] // 60)}:{int(avg_stats['last_boarding_time'] % 60):02d}"
    
    return avg_stats

def plot_results(results):
    """Generate visualizations of simulation results."""
    # Create a figure with subplots
    fig = plt.figure(figsize=(15, 15))
    
    # Plot 1: Queue size over time
    ax1 = fig.add_subplot(3, 1, 1)
    queue_times, queue_sizes = zip(*results['queue_timeline'])
    base_time = datetime(2025, 2, 24, 8, 0)  # Using current date from system
    formatted_times = [base_time + timedelta(minutes=t - 8*60) for t in queue_times]
    
    ax1.plot(formatted_times, queue_sizes, 'b-')
    ax1.set_title('Queue Size Over Time')
    ax1.set_xlabel('Time')
    ax1.set_ylabel('Number of Workers')
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax1.set_xlim(base_time, base_time + timedelta(minutes=60))
    ax1.grid(True)
    
    # Highlight specific times
    for time_str in ["8:30", "8:45", "9:00"]:
        h, m = map(int, time_str.split(':'))
        highlight_time = base_time + timedelta(hours=h-8, minutes=m)
        queue_size = results['queue_sizes'][time_str]
        ax1.axvline(x=highlight_time, color='r', linestyle='--', alpha=0.7)
        ax1.text(highlight_time, max(queue_sizes)/2, f' {time_str}: {queue_size:.1f} workers', rotation=90, verticalalignment='center')
    
    # Plot 2: Number of walkers by floor
    ax2 = fig.add_subplot(3, 1, 2)
    floors = list(results['walkers_by_floor'].keys())
    walkers = [results['walkers_by_floor'][floor] for floor in floors]
    
    colors = ['#3498db', '#2ecc71', '#e74c3c']
    ax2.bar(floors, walkers, color=colors)
    ax2.set_title('Number of Workers Walking to Each Floor')
    ax2.set_xlabel('Floor')
    ax2.set_ylabel('Number of Workers')
    ax2.set_xticks(floors)
    ax2.set_ylim(0, max(walkers) * 1.2)  # Add some space above the tallest bar
    
    # Add values on top of bars
    for i, v in enumerate(walkers):
        ax2.text(floors[i], v + 0.5, f'{v:.1f}', ha='center')
    
    # Plot 3: Wait time by arrival time
    ax3 = fig.add_subplot(3, 1, 3)
    arrival_times, wait_times = zip(*results['waiting_time_by_arrival'])
    base_time = datetime(2025, 2, 24, 8, 0)
    arrival_times_formatted = [base_time + timedelta(minutes=t - 8*60) for t in arrival_times]
    
    ax3.scatter(arrival_times_formatted, wait_times, alpha=0.5, s=20, c='purple')
    ax3.set_title('Wait Time by Arrival Time')
    ax3.set_xlabel('Arrival Time')
    ax3.set_ylabel('Wait Time (minutes)')
    ax3.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    ax3.set_xlim(base_time, base_time + timedelta(minutes=60))
    ax3.grid(True)
    
    # Add average wait time line
    avg_wait = results['average_wait_time']
    ax3.axhline(y=avg_wait, color='r', linestyle='-', label=f'Average: {avg_wait:.2f} min')
    ax3.legend()
    
    plt.tight_layout()
    return fig

# Run simulations and print results
def main():
    np.random.seed(42)  # For reproducibility
    num_simulations = 1000
    print(f"Running {num_simulations} simulations...")
    results = run_multiple_simulations(num_simulations)
    
    print("\nSimulation Results:")
    print(f"Average wait time: {results['average_wait_time']:.2f} minutes")
    print("\nNumber of people who walked to each floor:")
    for floor, count in results['walkers_by_floor'].items():
        print(f"  Floor {floor}: {count:.1f} workers")
    
    print(f"\nTime the last worker boards the elevator: {results['last_boarding_time']}")
    
    print("\nAverage queue size at specific times:")
    for time, size in results['queue_sizes'].items():
        print(f"  {time}: {size:.1f} workers")
    
    # Create visualizations
    fig = plot_results(results)
    plt.savefig("elevator_simulation_results.png", dpi=300, bbox_inches='tight')
    plt.show()
    
    return results

if __name__ == "__main__":
    main()
```
