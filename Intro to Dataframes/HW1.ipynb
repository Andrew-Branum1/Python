```python
# create dataframe from exteral files
# you can also read types of files, such as .txt, .csv, .json
intro_df =  pd.read_csv('apple_quality.csv', sep=",", header=None)
intro_df.columns = ["a_id", "size", "weight", "sweetness", "crunchiness", "juiciness", "ripeness", "acidity", "quality"]
#intro_df.head() # show only the first 5 rows
intro_df
```

```python
## Plot boxplots of features
plt.figure(figsize=(10, 5))
intro_df[["size", "weight", "sweetness", "crunchiness", "juiciness", "ripeness", "acidity", "quality"]].boxplot()
plt.title('Box plots of the features')
plt.xlabel('Feature')
plt.ylabel('Value Scale')
plt.show()
```

```python
# Define the columns for which you want to remove outliers
columns_to_process = ['size', 'weight', 'sweetness', 'crunchiness', 'ripeness', 'juiciness', 'acidity']

# Dictionary to store the quartiles and interquartile ranges for each column
quartiles_iqr = {}

# Calculate quartiles and interquartile ranges for each column
for column in columns_to_process:
    p25 = intro_df[column].quantile(0.25)
    p75 = intro_df[column].quantile(0.75)
    iqr = p75 - p25
    quartiles_iqr[column] = (p25, p75, iqr)

# Remove outliers for each column and create a new DataFrame
new_df = intro_df.copy()
for column, (p25, p75, iqr) in quartiles_iqr.items():
    lowest_datum = p25 - 1.5 * iqr
    highest_datum = p75 + 1.5 * iqr
    new_df = new_df[(new_df[column] > lowest_datum) & (new_df[column] < highest_datum)]

# Display the new DataFrame without outliers
display(new_df)
```

```python
## Plot boxplots of features
plt.figure(figsize=(10, 5))
new_df[["size", "weight", "sweetness", "crunchiness", "juiciness", "ripeness", "acidity", "quality"]].boxplot()
plt.title('Box plots of the features')
plt.xlabel('Feature')
plt.ylabel('Values')
plt.show()
```
